#### Обработка "tbd" значений в рейтинге пользователей

Что говорит Google. `TBD = To Be Determined`.
Если у игры меньше чем 4 ревью, то рейтинг отображаться не будет:/. 

Пропущенных значений здесь и так хватает, а тут еще 3000 неизвестных значений. Возможно получится его вычислить исходя из рейтинга критиков?

Посмотрим, есть ли корреляция между `critic_score` и `user_score`. Других вариантов заменить tbd у нас нет, медиану или среднее мы поставить не можем (т.к. у нас  1/3 пропущенных значений). 

Если не получится, останется только вариант с NaN.

```python
# Создадим временный датафрейм без tbd
test = df.query('user_score_10 != "tbd"').astype({'user_score_10': 'float'})

print('Корреляция пользовательского рейтинга и рейтинга критиков:', test['user_score_10'].corr(test['critic_score_100']))
test.plot(x='user_score_10', y='critic_score_100',  kind='scatter', alpha=0.2);
```

 - Ох как красиво! Получается, **зависимость** все-таки есть, но `к линейному тренду` она стремится на высоких рейтингах.

* В этот раз мы делаем **преобразование данных** ради эксперимента уже после `блока исследования`. Поэтому есть возможность вернуться назад и сделать нужные корректировки, в зависимости от задачи.


Переместившись во времени, вспоминаем, что выбранный актуальный период `с 2012 по 2016 год`. Так ли все плохо в этом промежутке?

```python
print("Количество данных о годовых продажах за актуальный период")
(df
    .query('user_score_10 == "tbd" and release_year >= 2012')
    .release_year.value_counts().rename('tbd_count').to_frame()
    .join(df.release_year.value_counts().rename('total_count').to_frame()) 
    .assign(
        tbd_percentage = lambda x: x['total_count'] / x['tbd_count'] #Тренируемся
    )
)
```

Не так много данных tbd как до 2012 года, но все же 30%. Наверное, придется все-таки `заполнить NaN-ами`.


Зависимость продаж будем оценивать по оставшися данным.

```python
df.user_score_10 = df.user_score_10.replace("tbd", np.NaN)
df.user_score_10 = pd.to_numeric(df.user_score_10)

df.dtypes
```

### Пропущенные данные

Мы уже успели увидеть общую картину по пропуском в блоке знакомства с данными.
Посмотрим еще раз на наш пласт работы.

```python
def print_missing_values(data):
    ''' Печатает количество и долю пропусков в одной таблице
    и возвращает сгенерированную таблицу
    '''    
    d = (data
         .isna().sum().rename('missing_count').to_frame()
         .join(df.isna().mean().rename('missing_ratio').to_frame())  
         .sort_values(by='missing_count', ascending=False)
        )
    display(d)
    return(d)
    
print_missing_values(df).head(7)
```

#### Все рейтинги

Делаем предположение, что в большинстве случаев мы имеем в рейтингах **связанные пропуски** (пользовательский, критиков и ESRB).

- Проверим, для этого создадим суммирующий столбец `sum_score` для рейтинга. 
- Если `user_score`, `critic_score` = NaN, то в сумме мы тоже **получим 0**.

```python
df['sum_score'] = df[['critic_score_100', 'user_score_10']].sum(axis=1) # axis=1 т.к. сумма по столбцам
df.head(5)
```

Все работает, проверяем.

### Итоговый пайплайн и дополнительные параметры

Сформируем по итогам первого блока пайплайн для предобработки датасета.

Такое учебное задание, снова берем исходный датафрейм `data`, но **все сделанные выше изменения запишем одним шагом**.
Задача из серии вспомнить все, бывает полезно.

Пайплайн будет очень короткий, рассуждений и аналитики у нас было очень много, а преобразований датасета - мало.

```python
def show_df(df):
    '''Принт датафрейма для пайплана. Промежуточный результат'''
    df = df.copy()
    display(df.head(5))
    return df

def show_info(df):
    '''Вывод основных характериристик датафрейма
    для оценки преобразований в пайплайне предобработки данных
    '''
    print("***Основные характеристики датафрейма***")
    display(df.columns.to_list)
    
    print("размер датасета:", df.shape[0])
    
    # используем функцию, созданную выше
    print_missing_values(df)
    
    return df
    
def rename_all_cols(df):
    ''' Меняем ВСЕ названия столбцов'''
    df = df.copy()
    df.columns = ['name', 'platform', 'release_year', 'genre', 'NA_sales_mln', 'EU_sales_mln',  'JP_sales_mln', 'other_sales_mln', 'critic_score_100', 'user_score_10', 'rating_ESRB']

    # мы могли бы воспользоваться аттрибутом columns, но на выходе нам нужен датафрейм для пайпа
    return df

def modify_user_score(df):
    ''' Преобразуем данные в столбце с рейтингом пользователей'''
    df = df.copy()
    
    # Заменяем tbd на пропуски
    df.user_score_10 = df.user_score_10.replace("tbd", np.NaN)
    
    # Корректируем тип столбца на вещественный
    df.user_score_10 = pd.to_numeric(df.user_score_10)
    return df
    
df = (
    # исходный датафрейм
    data
    
    # смотрим исходный датасет
    .pipe(show_df)
    
    # и его характеристики
    .pipe(show_info)
    
    # переименовываем столбцы
    .pipe(rename_all_cols)
    
    # меняем формат столбца с годом выпуска
    .astype({'release_year': 'Int64'})   
    
    # преобразуем формат данных в рейтинге пользователей
    .pipe(modify_user_score)
    
    # удаляем строки с пропущенным названием игр (и автоматом - жанров)
    .dropna(subset=['name'])  
     
          
    # дополнительный последний шаг - добавляем столбец с суммарными продажами
    .assign(
        name =  lambda x: x.name.str.lower(),
        genre = lambda x: x.genre.str.lower(),

        total_sales_mln = lambda x:  x[['NA_sales_mln','EU_sales_mln','JP_sales_mln','other_sales_mln']].sum(axis=1)
    )  
 
)

# Смотрим полученный датасет и его характеристики
show_info(df)
df.head()
```

**Вывод**

Мы хорошо попрактиковались. Видно, что в полученном датасете:
- нет 2-х удаленных дубликатов
- названия игр и жанров в нижнем регистре 
- изменены названия стоблцов
- `user_score_10` и `release_year` получили другой тип данных

**Временной отрезок** для анализа периода жизни игр мы выбрали.

Теперь нам нужно выделить **ТОП- платформы** с наибольшими суммарными продажами и посмотреть, как менялись их **продажи в разрезе лет.**


- Возьмем для начала платформы с суммарным количеством проданных `копий игр > 600`.
- Если данных не будет хватать, посмотрим платформы с более `низкими объемами > 200`.

```python
# Графики с количеством проданных копий в год для ТОП-платформ"

# Сразу обернем в функцию
# Регулируем мин.год для среза и мин.объем сумм.продаж для платформ
def plot_platforms_life (data, min_year, min_games_sale):

    # Учимся делать переносы строк и вложенные группировки
    # Конечно лучше группировку сохранить в отдельной переменной
    (
        data
        #оставляем временной отрезок, который определили выше
        .query('release_year >= @min_year')
        [data.platform.isin 
        # отбираем для графика только те платформы, у которых количество продаж > 600
        # имена этих игр - индексы в нашей группировке
            (
                data.groupby('platform')    
                .agg(games_sale_mln = ('total_sales_mln','sum'))
                .loc[lambda x: x['games_sale_mln'] >= min_games_sale] # Учимся фильтровать в пайпе
                .index
            )]
        .pivot_table(index='release_year', columns = 'platform', values='total_sales_mln', aggfunc='sum')
        .pipe(show_df) # смотрим результат в виде таблицы
        .plot(style='o-', grid=True, figsize=(15,8))        
    );
    
    plt.title(label=f"Миллионы проданных игр в год на ТОП- платформах (общее количество продаж > {min_games_sale} млн)")
    plt.ylabel("Млн проданных игр")
    plt.show()

plot_platforms_life (df, 1999, 600)
```

Добавим теперь менее топовые платформы, какие все-таки там тренды. Ведь не факт, что такие легенды, как здесь, появятся снова))

### Влияние рейтингов на продажи платформ (все регионы)

Для оценки корреляции между рейтингом и продажами возьмем для начала одного подопытного кролика.
Им будет популярная платформа PS4. Наверняка ребята хорошо вложились в маркетинг:)

Отдельно рассмотрим отзывы пользователей и критиков.

```python
# Сразу оборачиваем в функцию
def score_sales_corr(data, platform):
    dt = data.query('platform == @platform')
   
    print(f"Корреляция рейтинга пользователей и годовых продаж игр на {platform}:", 
          dt['user_score_10'].corr(df.total_sales_mln))
    
    print(f"Корреляция рейтинга критиков и годовых продаж игр на на {platform}:", 
          dt['critic_score_100'].corr(df.total_sales_mln))    
    
    fig, axes = plt.subplots(1,2, figsize=(12,4))
    dt.plot(ax = axes[0], x='user_score_10', y='total_sales_mln',  kind='scatter')
    dt.plot(ax = axes[1], x='critic_score_100', y='total_sales_mln',  kind='scatter')
    
    axes[0].set_title(label = f"График корреляции рейтинга ПОЛЬЗОВАТЕЛЕЙ и продаж")
    axes[1].set_title(label = f"График корреляции рейтинга КРИТИКОВ и продаж")
    plt.show()
    

score_sales_corr(df, "PS4")   
```

 - В случае с рейтингом критиков можно проследить определенный тренд. Корреляция, конечно, небольшая
 - Видно,что взаимосвязь усиливается для рейтингов выше 50
 - По пользовательским рейтингам ничего интересного найти не получается
 
 А как объстоят дела у других платформ?
 

```python
for platform in ["XOne", "3DS", "PC"]:
    score_sales_corr(df, platform)
```

Интересно, а будет ли разница, если мы изменим размерность critic_score_100 для PC? Cейчас размерность продаж примерно 0-12 млн. Попробуем.

```python
score_sales_corr(df.assign(critic_score_100 = lambda x: x['critic_score_100'] / 10),
                "PC")
```

Картинка осталась та же, и значения сохранились.

### Получаем новый датасет на основе регионов

```python
df.head(3)
```

В текущем блоке нам будет не очень удобно работать с данными в таком формате, объединим продажи в разных регионах в одну категорию!

- Добавим столбец-категорию `region _name`, который будет принимать значения  `америка`, `европа`, `япония`, `другие`
- Добавим столбец `sales_mln`, где будут отображаться продажи в выбранном регионе (cюда попадут значения в столбцах `NA_sales_mln`, `EU_sales_mln`, `JP_sales_mln`, `other_sales_mln`)

Мы транспонируем таблицу, **каждая строка будет отражать продажи в отдельно взятом регионе**.
Т.е. как бы из формата pivot_table перейдем в формат c возможностями группировок.

```python
# Готовим для себя самопроверку результатов предобразования
print(f"Количество данных по продажам в исходном датасете: {df.shape[0]}")
print(f"После разделения 4-х типов регионов по отдельным строкам должны получить: {df.shape[0]*4}")
```

```python
# Создаем новый датафрейм на базе данных о продажах в разных регионах

df_reg = (df
        # переименуем сразу столбцы в нужные нам названия внутри категории
        .rename(columns={'NA_sales_mln':'америка', 'EU_sales_mln':'европа', 'JP_sales_mln':'япония', 'other_sales_mln':'другие'})
        
        # делаем unpivot данных о продажах разных регионов
        # имена нужных колонки из df записываем в id_vars
        .melt(id_vars=['name', 'platform', 'release_year', 'genre', 'rating_ESRB'], 
            value_vars=['европа', 'америка', 'япония', 'другие'],
            var_name='region_name', value_name='sales_mln')
         )

# Проверяем, что ничего не потеряли после преобразований
print("Новый датасет `df_reg` с категорией продаж по регионам в столбце `region_name`:")
display(df_reg.sort_values(by=['name', 'platform']).head(5))
print("Объем данных после преобразований:", df_reg.shape[0])

df.sort_values(by=['name', 'platform']).head(5)
```

- Количество данных в новом датасете корректное
- Данные по продажам со старым датасетом совпали

### ТОП популярных платформ 

Для оценки популярности посчитаем **доли продаж** по платформам внутри каждого региона.

```python
# Сразу обернем в функцию, пригодится для исследования жанров и возможно других проектов
def sales_ratio (data, index_categ, column_categ):
    ''' Cоздание pivot_table c долями продаж по каждой категории
    
    index_categ -  категория, которую изучаем. По ней будет индексирована таблица
    column_categ - категория, внутри которой будем считать доли продаж '''
    
    ratio_df = data.copy()
    # Строим сводную таблицу с суммарными продажами по каждой категории index_categ
    # Добавляем margins - суммирующие столбцы и строки
    ratio_df = (data
             .pivot_table(index=index_categ, columns=column_categ, values='sales_mln', aggfunc='sum', margins=True)         
    )

    # Вычисляем доли продаж внутри каждой категории column_categ
    # Используем для этого margins All - сумму по каждой колонке = общее количество продаж в column_categ
    ratio_df = ratio_df/ratio_df.loc["All"]

    # Убираем из итоговой таблицы суммирующий столбец и строку 
    # заодно вспомним забытый в первых модулях iloc
    ratio_df = ratio_df.iloc[:-1 , :-1]
    
    return(ratio_df)

# Строим сводную таблицу с долями продаж по каждой платформе внутри регионов
sales_regions = sales_ratio(df_reg, 'platform', 'region_name')

# Выводим таблицу
cm = sns.light_palette("orange", as_cmap=True)
sales_regions.style.background_gradient(cmap=cm)
```

#### Анализ распределений пользовательских рейтингов  

Итак, это наш первый шаг  - сравнение выбранных платформ на основе **средних пользовательских рейтингов** с помощью статистических гипотез.

- В п.4.4. мы уже определили для себя **ТОП-популярных платформ**. Возьмем  в исследование `Xbox One` и `PC`.

- Помним из п.3.2 , что средний **уровень продаж** игр на этих платформах сильно различался.

- Сохранится ли такой же разрыв на уровне пользовательских отзывов?

Прежде чем переходить к гипотезам, посмотрим **распределения рейтингов** на гистограммах (внутри платформ) и сделаем предварительные выводы

```python
# Заведем в функцию построение таких графиков, т.к. часто используем 
def plot_hist_2categ(data, column, categ1, categ2, value, 
                     bins, legend_list, title, xlabel, ylabel):
    '''Строит распределение величины value на одном графике
    для двух категорий categ1 и сateg2 в столбце column'''
    
    fig, axes = plt.subplots(figsize=(8,4))

    data[data[column] == categ1][value].hist(bins=bins)
    data[data[column] == categ2][value].hist(bins=bins, alpha=0.7)

    axes.legend(legend_list, fontsize=12)
    plt.title(label=title, fontsize=16)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.show()

plot_hist_2categ(df, "platform", "XOne", "PC", "user_score_10", 
                    bins=15, legend_list=["Игры на XOne", "Игры на PC"], 
                    title="Распределение пользовательских рейтингов игр", 
                    xlabel="Пользовательский рейтинг", 
                    ylabel="Количество отзывов"
                )

# Вывод выборочных статистик для распределений также обернем в функцию
def statistics_for_ttest(data, column, categ1, categ2, value):
    display(
         data [(data[column] == categ1) | (data[column] == categ2)]        
         .groupby(column)[value]
         .agg(['count', 'mean', 'median', 'std', 'var'])
        )
print("Cтатистики распределений пользовательских рейтингов для XOne и PC:")
statistics_for_ttest(df, "platform", "XOne", "PC", "user_score_10")
```

- **Объем данных** в выборках **достаточно мал**. Значит, для получения хороших результатов нам стоит уделить особое внимание параметрам в стат.тесте

#### Построение статистических гипотез

Исходя из выводов выше мы хотели бы проверить гипотезу, что `cредние пользовательские рейтинги платформ Xbox One и PC 
одинаковые`.


1.Сформулируем **нулевую гипотезу H0**: *Cредние пользовательские рейтинги платформ Xbox One и PC равны*

2.**Альтернативная гипотеза H1** будет двухсторонней: *Cредние пользовательские рейтинги платформ Xbox One и PC не равны*

Для проверки гипотезы равенства среднего ЦПТ позволяет нам воспользоваться **критерием Стьюдента для независимых выборок**.
Этот же критерий мы применим в п. 5.2 для сравнения двух популярных жанров.

Здесь снова библиотека scipy нам в помощь и метод `scipy.stats.ttest_ind (array1, array2,equal_var)`

С параметром `equal_var` в этот раз не все так очевидно, как в прошлом проекте.
- Во-первых, выборки очень маленькие
- Во-вторых сложнее сделать предположение о дисперсиях ген.совокупностей (хотя выборочные дисперсии различаются)

Попробуем **обосновать разницу в дисперсиях** в ген.сов. тем, что у XOne намного выше продажи и в целом платформа более современная, популярная. Значит, у нее д.б. не такая сильная вариативность в низких оценках, как у PC.

```python
# Функцию взяли из прошлого проекта, очень удобно
# Добавили несколько выводов о параметрах теста
def check_ttest_pvalue (distr1, distr2, alpha, var_equal):    
    ''' Вывод результатов теста Стьюдента для равенства средних 2-х выборок '''
    
    print("Тест Стъюдента")
    print("предположение о равенстве дисперсий:", var_equal)
    print(f"выбранный уровень значимости: {alpha} \n")
    
    test_result = st.ttest_ind(distr1, distr2, equal_var=var_equal)
    print('p-значение: ', test_result.pvalue.round(7))
    
    # критический уровень статистической значимости
    # если p-value окажется меньше него - отвегнем гипотезу
    if test_result.pvalue < alpha:
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу") 
        
        
# var_equal=False, т.к. предположение, что дисперсии разные
# alpha взяли .05
# (!!!) исключаем пропуски в данных о рейтинге (п. 2.2.1)
check_ttest_pvalue(
                df[df['platform'] == "XOne"].user_score_10.dropna(),
                df[df['platform'] == "PC"].user_score_10.dropna(),
                .05, False
)
```

**Вывод**

- Если средние пользовательские рейтинги платформ `Xbox One` и `PC` были бы равны, то мы бы получили картинку выше (распределения выборок) **с вероятностью 0.549**.


- Вероятность достаточно высокая, и при заданном уровне значимости alpha=0.5 мы не можем опровергнуть нулевую гипотезу

Таким образом, мы можем предполагать, что несмотря на `сильную разницу в суммарных продажах`, 2 наши популярные платформы потенциально могут иметь `одинаковый средний рейтинг пользователей` (опровергнуть обратное нам не удалось:)
